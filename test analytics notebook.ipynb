```python
import os
import pandas as pd
import numpy as np
```

```python
daily_fp = os.path.join('daily_sales.csv')
monthly_fp = os.path.join('monthly_sales.csv')
```

```python
daily_sales = pd.read_csv(daily_fp)

monthly_sales = pd.read_csv(monthly_fp)
monthly_sales.set_index('Month, Year')
mean_sales_DOW = daily_sales.groupby('Day of Week').mean()

daily_sales['Day'] = pd.to_datetime(daily_sales['Day'])
```

```python
mean_sales_DOW.to_csv('mean_sales_DOW.csv')
monthly_sales = monthly_sales.set_index('Month, Year')
```

```python
monthly_NE = monthly_sales.iloc[:, monthly_sales.columns.str.contains('NE')]
```

```python
monthly_sales[monthly_sales.columns[monthly_sales.columns.str.contains('NE')]]


monthly_sales.columns[monthly_sales.columns.str.contains('NE')]
```

```python
def divide(boom):
    return boom / monthly_NE.sum(axis = 1)
```

```python
monthly_NE.apply(divide).T.to_csv('NE_percentage_sales.csv')
```

```python
CF_monthly = monthly_sales[monthly_sales.columns[monthly_sales.columns.str.contains('CF')]]
FF_monthly = monthly_sales[monthly_sales.columns[monthly_sales.columns.str.contains('FF')]]
HM_monthly = monthly_sales[monthly_sales.columns[monthly_sales.columns.str.contains('HM')]]
```

```python
HM_monthly.T.sum().to_csv('agg_HM_monthly.csv')
CF_monthly.T.sum().to_csv('agg_CF_monthly.csv')
FF_monthly.T.sum().to_csv('agg_FF_monthly.csv')
```

```python

def cool(month):
    if 'Jan' in month:
        return 'Winter'
    elif 'Feb' in month:
        return 'Winter'
    elif 'Mar' in month:
        return 'Spring'
    elif 'Apr' in month:
        return 'Spring'
    elif 'May' in month:
        return 'Spring'
    elif 'Jun' in month:
        return 'Summer'
    elif 'Jul' in month:
        return 'Summer'
    elif 'Aug' in month:
        return 'Summer'
    elif 'Sep' in month:
        return 'Fall'
    elif 'Oct' in month:
        return 'Fall'
    elif 'Nov' in month:
        return 'Fall'
    else:
        return 'Winter'
            
    
```

```python
monthly_sales['Season'] = monthly_sales['Month, Year'].astype(str).apply(cool)
```

```python
mean_mon_by_season = monthly_sales.groupby('Season').mean()
```

```python
count_mon = monthly_sales.groupby('Season').count()
```
